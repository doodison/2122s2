## Reading and writing files

```python
#reading txt
wordlist=[]
```

```python
with open('Example_word_list.txt', mode = 'r', encoding='utf-8') as f:
    for line in f:
        wordlist.append(line.strip())
```

```python
#Changing txt
```

```python
#Writing csv
with open('newwordlist.txt', mode = 'w', encoding='utf-8') as f:
    f.writelines(wordlist)
```
